# just go straight to the searh results page for Habitat for Humanity and Indiana

```python
dr.get('https://projects.propublica.org/nonprofits/search?utf8=%E2%9C%93&q=Habitat+for+Humanity&state%5Bid%5D=IN&ntee%5Bid%5D=&c_code%5Bid%5D=')
```

```python
soup = BeautifulSoup(dr.page_source, 'html.parser')
```

```python
soup
```

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
portion_link_list = []
for link in soup.find_all('a'):
    portion_link_list.append(link.get('href'))
```

```python
str(portion_link_list)
```

# find only the partial links needed

```python
mylist = re.findall(r'(/nonprofits.organizations/\d+)', str(portion_link_list), re.DOTALL)  
```

## Check to see if it worked

```python
mylist[2]
```

```python
type(mylist[2])
```

# Convert those partial links into full links

```python
affiliate_links = []
for each in mylist:
    appendage = 'https://projects.propublica.org' + each
    affiliate_links.append(appendage)
```

```python
affiliate_links
```
