# 2014

```python
annual_affiliates_filings = []
large_frame = pd.read_csv('index_2014.csv', error_bad_lines = False)
org_frame = large_frame[large_frame['TAXPAYER_NAME'].str.contains(r'(?i)(Habitat for Humanity)')]
print('droping 990EZ and 990EO filing forms')
frame_after_drop1 = org_frame[org_frame.RETURN_TYPE != '990EO']
new_frame = frame_after_drop1[frame_after_drop1.RETURN_TYPE != '990EZ']
print("coverting data into a link")
for each in new_frame['OBJECT_ID']:
    appendage = 'https://s3.amazonaws.com/irs-form-990/' + str(each) + '_public.xml'
    annual_affiliates_filings.append(appendage)
```
