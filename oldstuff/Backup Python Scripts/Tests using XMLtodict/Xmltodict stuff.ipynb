# notice the class collections.OrderedDict
We can use the different data types of the value to know if the node has any children or if it is a leaf. I hypothesised that all values with dictionaries contained children and all other values were of leaf nodes. This is, for the vast majority, true.

Below is a function that 'dives' into a dictionary tree and converts the structure into super easy to read text output. The magic of the function is recursion. 

From preliminary runs of various versions of the code below, I found some odditites. For instance, a list where there should be a dictionary. Also a list that is the value. Additionaly there can be nodes that contain no values.

```python
def scan_dive(node):
    for item in node.items():
        if item[1] != None:
            if type(item[1]) == str:
                print(item[0] + " is a leaf node")
            
            elif type(item[1]) == list:
                print(item[0] + ' is a list... unpacking and exploring the sublevel.\n')
                try:
                    scan_dive(item[1][0])
                    print('\nEnd of sub dictionary ' + item[0] + ', moving up a level.\n')
                except AttributeError:
                    print('\nOh! This does not contain any subdictionaries. ' + item[0] +' is actually leaf node')
            
            else:
                print(item[0] + 'is a dictionary... exploring the sublevel.\n')
                scan_dive(item[1])
                print('\nEnd of sub dictionary ' + item[0] + ', moving up a level.\n')
        else:
            print('Found a Nonetype situation at ' + item[0])
```

```python
scan_dive(data['Return'])
```

# using same logic structure we can also count up the number of leaves on the tree.



```python
def node_count(node):
    count = 0
    for item in node.items():
        if item[1] != None:
            if type(item[1]) == str:
                count += 1
            
            elif type(item[1]) == list:
                try:
                    count += node_count(item[1][0])
                except AttributeError:
                    pass
            else:
                count += node_count(item[1])
    return count
```

```python
node_count(data)
```

# Building out the flat database.
Currently the structure of imbedded dictionaries is annoying. Long term goal is to generate a flat data base. A flat dictionary would help here. Or at the very least, provide a list of the path names for all the nodes. 

The logic will be similiar to logic found earlier.

```python
def nodenames_constructor(parent):
    node_names = []
    
    for item in parent.items():
        if item[1] != None:
            if type(item[1]) == str:
                node_names.append(item[0])
            
            elif type(item[1]) == list:
                try:
                    childlist_nodenames = nodenames_constructor(item[1][0])
                    for child in childlist_nodenames:
                        string_name = item[0] + '.' + child
                        node_names.append(string_name)
                except AttributeError:
                    print("encountered an error")
            
            else:
                child_nodenames = nodenames_constructor(item[1])
                for child in child_nodenames:
                    string_name = item[0] + '.' + child
                    node_names.append(string_name)
            
    return node_names



#def pathname_1_time_constructor(parent):
#    path_names = []
#    for item in parent.items():
#        if item[1] != None:
#            if type(item[1]) == str:
#                path_names.append(item[0])
#            
#    return path_names
```

```python
def node_values(parent):
    value_list = []
    for item in parent.items():
        if item[1] != None:
            if type(item[1]) == str:
                value_list.append(item[1])
            
            elif type(item[1]) == list:
                try:
                    child_values = node_values(item[1][0])
                    for value in child_values:
                        value_list.append(value)
                except AttributeError:
                    print('encountered an error')
            
            else:
                child_values = node_values(item[1])
                for value in child_values:
                    value_list.append(value)
    
    return value_list
            
```
